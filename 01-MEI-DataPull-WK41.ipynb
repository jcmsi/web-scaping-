```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time 
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.microcenter.com/search/search_results.aspx?N=4294967288+4294818187&NTK=all&sortby=numreviews&rpp=96")
js_content = driver.page_source
```

```python
soup = BeautifulSoup(js_content, "html.parser")
```

```python
quote = soup.find("div", class_="normal")
print(quote)
```

## Name

```python
page = 1
name = []
while page != 3:
        time.sleep(20)
        soup = BeautifulSoup(js_content, "html.parser")
        for p in soup.find_all('div', class_='normal'):
            name.append(p.get_text(strip=True))
        page = page + 1
```

```python
len(name)
```

```python
name
```

```python
import pandas as pd
df = pd.DataFrame(name, columns=['name'])
df
```

# SKU

```python
page = 1
SKU = []
while page != 3:
        time.sleep(20)
        soup = BeautifulSoup(js_content, "html.parser")
        for p in soup.find_all('p', class_='sku'):
            SKU.append(p.get_text(strip=True))
        page = page + 1
```

```python
len(SKU)
```

```python
print(SKU)
```

```python
df['SKU'] = pd.Series(SKU)
```

```python
df.head()
```

```python
df['SKU']=df['SKU'].str.slice(start=4)
```

```python
df.head()
```

# Price

```python
page = 1
price1 = []
while page != 3:
        time.sleep(20)
        soup = BeautifulSoup(js_content, "html.parser")
        for p in soup.find_all('span', itemprop='price'):
            price1.append(p.get_text(strip=True))
        page = page + 1
```

```python
len(price1)
```

```python
print(price1)
```

```python
df['price1'] = pd.Series(price1)
```

```python
df['price']=df['price1'].str.slice(start=9)

df.head()
```

```python
df = df.drop(columns=['price1'])
```

```python
df
```

# Discount

```python
page = 1
discount1 = []
while page != 3:
        time.sleep(20)
        soup = BeautifulSoup(js_content, "html.parser")
        for p in soup.find_all('div', class_='price'):
            discount1.append(p.get_text(strip=True))
        page = page + 1
```

```python
df2 = df1[df1['discount1'].str.contains('Original')]
```

```python
df2['discount1'] = df2['discount1'].str.slice(start=27)


df2
```

```python
df2['discount1'] = df2['discount1'].str[:8]
df2
```

# In-Stock

```python
page = 1
instock = []
while page != 3:
        time.sleep(20)
        soup = BeautifulSoup(js_content, "html.parser")
        for p in soup.find_all('span', class_='inventoryCnt'):
            instock.append(p.get_text(strip=True))
        page = page + 1
```
